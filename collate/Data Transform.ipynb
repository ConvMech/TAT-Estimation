```python
bool_arry = [data['puborders']['sneaky'] == 'f', data['puborders']['order_state'] == 'complete']
#print (bool_arry)
res = np.all(bool_arry, axis = 0)
puborders = data['puborders'].loc[res,
                            ['order_id', 'order_created_at', 'priority', 'order_complexity', 
                             'order_image_score', 'order_deliverable', 'site_id', 'number_of_stories',
                             'complexity_attributes_total_roof_facets', 'number_of_images', 'average_image_score', 
                             'image_count_1_side', 'image_count_2_sides', 'turn_around_time_first_completion']]
puborders["order_created_at"] = pd.to_datetime(puborders["order_created_at"], utc=True, errors='coerce')
puborders = puborders[puborders["order_created_at"].notnull()]
print(len(puborders))
display(puborders)
```

```python
'''
puborders.at[puborders['order_complexity'] == 'Average' , 'order_complexity'] = float32(1.0)
puborders.at[puborders['order_complexity'] == 'Simple' , 'order_complexity'] = float32(0.0)
puborders.at[puborders['order_complexity'] == 'Complex Residential' , 'order_complexity'] = float32(2.0)
puborders.at[puborders['order_complexity'] == 'Custom' , 'order_complexity'] = float32(3.0)

puborders.at[puborders['order_deliverable'] == 'roof' , 'order_deliverable'] = float32(0.0)
puborders.at[puborders['order_deliverable'] == 'complete' , 'order_deliverable'] = float32(1.0)
'''

def fn1(x):
    if x == 'Simple':
        return 0.0
    elif x == 'Average':
        return 1.0
    elif x == 'Complex Residential':
        return 2.0
    elif x == 'Custom':
        return 3.0
def fn2(x):
    if x == 'roof':
        return 0.0
    elif x == 'complete':
        return 1.0
puborders['order_complexity'] = puborders['order_complexity'].apply(fn1)
puborders['order_deliverable'] = puborders['order_deliverable'].apply(fn2)

display(puborders)
```

```python
# process date and order id, convert to number
                                                                    
puborders['order_created_at'] = puborders['order_created_at'].apply(lambda x: (x.dayofweek + \
                                                                               (x.hour + x.minute / 60.0 \
                                                                               ) / 24.0 ) / 7.0)
puborders['order_id'] = puborders['order_id'].apply(lambda x: int(x))
print(len(puborders))
puborders.dropna(inplace = True)
print(len(puborders))
```
