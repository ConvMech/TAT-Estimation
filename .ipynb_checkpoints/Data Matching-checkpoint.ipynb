```python
from functools import reduce
from datetime import datetime

def dictarr_to_df(dictarr):
    keys = list(reduce((lambda x, y: x.union(y)), [set(list(x.keys())) for x in dictarr]))
    
    z = pd.DataFrame(dictarr)
    z.index = z.time
    z.sort_index(ascending=True, inplace=True)

    for k in keys:
        z[k] = z[k].fillna(0).astype(int)

    return z
```

```python
filtered_orders_arr = list(filtered_orders.itertuples())
```

```python
def get_counts_for_datetime(the_datetime):
    before = order_counts_by_minute[order_counts_by_minute["time"] <= the_datetime].tail(1).iloc(0)[0]
    after  = order_counts_by_minute[order_counts_by_minute["time"] >= the_datetime].head(1).iloc(0)[0]

    if not before.empty and after.empty:
        return before

    if not after.empty and before.empty:
        return after
    
    before_int = int(time.mktime(before.time.timetuple()))
    after_int = int(time.mktime(after.time.timetuple()))
    
    # compare them and get the one closest to the actual time
    the_datetime_int = int(time.mktime(the_datetime.timetuple()))
    af_diff = abs(the_datetime_int - after_int)
    be_diff = abs(the_datetime_int - before_int)
    
    if af_diff <= be_diff:
        return after
    
    return before
    
def get_counts_for_order_row(order_row):
    return get_counts_for_datetime(order_row.order_created_at)
```

```python
print(len(filtered_orders_arr))
```

```python

print("Processing...")
start = time.time()


X_dict_arr = []
Y = []

i = 0
for order in filtered_orders_arr:
    counts = get_counts_for_order_row(order)
    md = dict(get_counts_for_order_row(filtered_orders_arr[0]))
    del md["time"]
    if order.order_deliverable == 'complete':
        md["deliverable_id"] = 3
    if order.order_deliverable == 'roof':
        md["deliverable_id"] = 2
     
    X_dict_arr.append(md)
    Y.append(int(order.turn_around_time_total))
    
    i+=1
    if i >= 1000:
        break


X = []
keys = list(X_dict_arr[0].keys())
for X_dict in X_dict_arr:
    datum = [ X_dict[k] for k in keys ]
    X.append(datum)


print("X: ", len(X))


print("Complete")
end = time.time()
print(end - start)
```
